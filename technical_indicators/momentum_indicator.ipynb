```python
import yfinance as yf
import pandas_ta as ta 
data=yf.download(tickers='SPY', period='max', interval='1d')
```

```python
data["RSI_10"] = ta.rsi(data["Close"], length=10)
```

```python
len(data)
```

```python
import plotly.graph_objects as go
df = data[-1000:-500]
fig = go.Figure(data=[go.Candlestick(x=df.index,
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'],
                name='OHLC'),
                go.Scatter(x=df.index, y=df['RSI_10'], name='RSI', yaxis='y2')
                ])
fig.update_layout(
    title='OHLC with RSI(10)',
    yaxis=dict(
        domain=[0.2, 1]
    ),
    yaxis2=dict(
        domain=[0, 0.2],
        anchor='free',
        overlaying='y',
        side='right',
        title='RSI'
    ))

fig.show()
```

```python
df.ta.stoch(high='High', low='Low', close='Close', fast_k=14, slow_k=3, slow_d=3, append=True)
```

```python
from plotly.subplots import make_subplots 

# subplots for indicators
fig = make_subplots(rows=2, cols=1)

# first plot candlestick
fig.add_trace(go.Candlestick(x=df.index,
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close']),
                row=1, col=1)

# second plot ADX
fig.add_trace(go.Scatter(x=df.index, y=df['STOCHk_14_3_3'], name='s1'), row=2, col=1)
fig.add_trace(go.Scatter(x=df.index, y=df['STOCHd_14_3_3'], name='s2'), row=2, col=1)
fig.update_layout(
    xaxis=dict(rangeslider=dict(visible=False)),
    xaxis2=dict(rangeslider=dict(visible=False))
)

fig.show()
```
